# **Correlated Wishart Orthogonal Ensemble (CWOE)**

```python
'''
Call the class with one of the following correlation definition:
 * Pearson: 'p', 'pearson', 'P'
 * Spearman Rank: 's', 'spearman', 'S'
 * Distance: 'd', 'distance', 'D'
'''
correlation_coeff = 'pearson'

'''Read the moments and extract the acc values'''
if correlation_coeff == 'pearson' or correlation_coeff == 'p' or correlation_coeff == 'P':
    df_moments = pd.read_csv('../data/processed/pcc/pcc_moments.csv',header=0)
elif correlation_coeff == 'spearman' or correlation_coeff == 's' or correlation_coeff == 'S':
    df_moments = pd.read_csv('../data/processed/srcc/srcc_moments.csv',header=0)
elif correlation_coeff == 'distance' or correlation_coeff == 'd' or correlation_coeff == 'D':
    df_moments = pd.read_csv('../data/processed/dcc/dcc_moments.csv',header=0)

acc = np.array(df_moments['ACC'])
```

```python
'''Set a random seed just to maintain consistency'''
np.random.seed(3)

'''Call the CWOE method'''
T.ensemble(acc, type='CWOE')
```
